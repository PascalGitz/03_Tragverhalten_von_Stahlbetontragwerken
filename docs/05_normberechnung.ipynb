```python
import matplotlib.pyplot as plt
```

```python
plt.rcParams.keys()
```

```python
def faktor_w1(rho, rhobar, d, h, phi):
    return (1- 20*rhobar) / (rho**0.7*10) *(0.75+0.1*phi) *(h/d)**3
```

```python
rho_list = np.linspace(0.01/100, 2/100, 20000)
rho_bar_list = [0, 0.5/100, 1/100]
h = 1
d = 0.85
phi = 1

f_g_list = [[], [], []]
for rho in rho_list:
    for i,rho_bar in enumerate(rho_bar_list):
        f_g_list[i].append(faktor_w1(rho, rho_bar, d, h, phi))


colors = ['black', 'black', 'black']
linestyles = ['-', '--', 'dotted']
label = [r'$\rho^,$ = 0 \%', r'$\rho^,$ = 0.5 \%', r'$\rho^,$ = 1 \%']
for j, f_g in enumerate(f_g_list):
    plt.plot(rho_list*100, f_g, linestyle = linestyles[j], color = colors[j], label = label[j])

plt.xlabel(r'$\rho$ [\%]')
plt.ylabel(r'$f_g$ [-]')
plt.ylim(0, 18)
plt.legend()

plt.savefig('images/f_g.svg')


plt.show()
```
